```python
%matplotlib inline
import pandas as pd
import numpy as np
from datetime import datetime

# Note: Here I'm loading my movie csv file to make a data frame
# You can work on your own data frame

movies = pd.read_csv('2013_movies.csv', parse_dates={'Date':['ReleaseDate']})

movies.head()
```

### Predict Gross using Budget only

```python
movies_with_budget = movies[ -np.isnan(movies.Budget) ]
movies_with_budget.tail()
```

```python
movies_with_budget = movies[ -np.isnan(movies.Budget) ]
```

```python
movies_with_budget['Ones'] = 1.0
movies_with_budget.tail()
```

```python
#adding log
domestic_total_gross_log = np.log(movies_with_budget.DomesticTotalGross)
domestic_total_gross_log = pd.DataFrame(domestic_total_gross_log)
domestic_total_gross_log.rename(columns={'DomesticTotalGross': 'DomesticTotalGrossLog'}, inplace=True)

movies_with_budget = pd.concat([movies_with_budget,domestic_total_gross_log], axis=1)
movies_with_budget.tail()
```

```python
import statsmodels.formula.api as sm

X = movies_with_budget[['Budget', 'Ones']]
Y = movies_with_budget['DomesticTotalGross']
linmodel = sm.OLS(Y, X).fit()
linmodel.summary()
```

```python
#using log
X = movies_with_budget[['Budget', 'Ones']]
Y_log = movies_with_budget['DomesticTotalGrossLog']
linmodel_log = sm.OLS(Y_log, X).fit()
linmodel_log.summary()
```

#### Plot the model

```python
import matplotlib.pyplot as plt
plt.scatter(movies_with_budget.Budget, movies_with_budget.DomesticTotalGross)
plt.xlabel('Budget')
plt.ylabel('Box Office')
plt.plot(movies_with_budget.Budget, linmodel.predict(X), 'r')

from statsmodels.sandbox.regression.predstd import wls_prediction_std
prstd, conf_lower, conf_upper = wls_prediction_std(linmodel)
plt.plot(movies_with_budget['Budget'], conf_lower, 'y--')
plt.plot(movies_with_budget['Budget'], conf_upper, 'y--')
```

```python
#challenge 1
#using log
plt.scatter(movies_with_budget.Budget, movies_with_budget.DomesticTotalGrossLog, color='black')
plt.xlabel('Budget')
plt.ylabel('Box Office')
plt.plot(movies_with_budget.Budget, linmodel_log.predict(X), 'r')

from statsmodels.sandbox.regression.predstd import wls_prediction_std
prstd1, conf_lower1, conf_upper1 = wls_prediction_std(linmodel_log)
plt.plot(movies_with_budget['Budget'], conf_lower1, 'y--')
plt.plot(movies_with_budget['Budget'], conf_upper1, 'y--')
```

#### Plot the predictions of the model against true values

```python
# plot the predictions 
plt.plot(movies_with_budget['DomesticTotalGross'], linmodel.predict(X), 'go')
# plot the 1:1 actual gross line
plt.plot(movies_with_budget['DomesticTotalGross'], movies_with_budget['DomesticTotalGross'], 'r')
# axis labels
plt.xlabel('Actual Gross')
plt.ylabel('Predicted Gross by Rating Model')
```

```python
#using log
# plot the predictions 
plt.plot(movies_with_budget['DomesticTotalGrossLog'], linmodel_log.predict(X), 'go')
# plot the 1:1 actual gross line
plt.plot(movies_with_budget['DomesticTotalGrossLog'], movies_with_budget['DomesticTotalGrossLog'], 'b--')
# axis labels
plt.xlabel('Actual Gross')
plt.ylabel('Predicted Gross by Rating Model')
```

### Same regression with scikit.learn

```python
# Same model with scikitlearn

from sklearn.linear_model import LinearRegression

np_movies = np.matrix(movies_with_budget)

X = np_movies[:, 2]  # budget
Y = np_movies[:, 3]  # gross

sk_linmodel = LinearRegression()
sk_linmodel.fit(X,Y)

x, y = X.A1, Y.A1

print 'coefficient:\t %g'% sk_linmodel.coef_[0,0]
print 'intercept:\t %g'  % sk_linmodel.intercept_[0]
print 'R^2:      \t %g'  % sk_linmodel.score(X, Y)

plt.scatter(x, y)
plt.xlabel("Budget")
plt.ylabel("Box Office")
plt.plot(x, sk_linmodel.predict(X), 'r')
```

### Predict Gross using Ratings (categorical variable)

```python
m = movies_with_budget
#m.head()
pd.get_dummies(m['Rating'])
```

```python
m = pd.concat([m,pd.get_dummies(m['Rating'])], axis=1)
m.head()
```

```python
X = m[['PG', 'PG-13', 'R', 'Ones']]
Y = m['DomesticTotalGross']
rating_linmodel = sm.OLS(Y, X).fit()
rating_linmodel.summary()
```

#### Plot the predictions

```python
# plot the predictions 
plt.plot(m['DomesticTotalGross'], rating_linmodel.predict(X), 'go')
# plot the 1:1 actual gross line
plt.plot(m['DomesticTotalGross'], m['DomesticTotalGross'], 'b--')
# axis labels
plt.xlabel('Actual Gross')
#plt.ylabel('Predicted Gross by Rating Model')
```
